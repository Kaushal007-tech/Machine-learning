# Testing on trained model 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')

import nltk
```

```python
df = pd.read_csv('Product_review.csv')
print(df.shape)
df = df.iloc[200000:300000]
print(df.shape)
```

```python
df.head()
```

```python
test_data = df.drop(['asin','helpful','reviewTime','reviewerID','reviewerName','unixReviewTime','summary'],axis=1)
```

```python
test_data
```

```python
test_data.rename(columns = {'Unnamed: 0':'ID','overall':'Ratings'}, inplace = True)
```

```python
test_data
```

```python
test_data.isnull().sum()
```

```python
test_data["reviewText"].fillna("No data", inplace = True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

### Pre process the text of new date to give into model

```python
cv = CountVectorizer(max_features=100)
X = test_data["reviewText"]
X = cv.fit_transform(X).toarray()
```

## Load the model for prediction

```python
import joblib
classifire = joblib.load("random_forest_model")
```

```python
y_pred = classifire.predict(X)
```

```python
print(y_pred)
```

### Create another column of prediction 

```python
test_data["predicted_labels"] = y_pred.tolist()
test_data.head(10)
```
